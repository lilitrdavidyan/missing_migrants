```python
def create_unique_columns_dataframe(datasets):
    unique_columns = {}
    
    for idx, dataset in enumerate(datasets, start=1):
        for column in dataset.columns:
            if column not in unique_columns:
                unique_columns[column] = [0] * len(datasets)
            unique_columns[column][idx-1] = 1
    
    unique_columns_df = pd.DataFrame.from_dict(unique_columns, orient='index', columns=[f"Dataset {idx}" for idx in range(1, len(datasets)+1)])
    return unique_columns_df.sort_index(ascending=True)


create_unique_columns_dataframe([missing_migrants_df, missing_migrants_df_2, missing_migrants_df_3])
```

```python
column_mapping = {
    'Location Coordinates': 'Coordinates',
    'Incident year': 'Year',
    'Reported Year': 'Year',
    'Information Source': 'Source',
    'Location Description': 'Location of death',
    'Migrantion route': 'Migration route',
    'Migration Route': 'Migration route',
    'Number of Dead': 'Number Dead',
    'Region': 'Region of Incident',
    'Total Number of Dead and Missing': 'Total Dead and Missing'
}
```

```python
def rename_columns(dataset, column_mapping):
    renamed_dataset = dataset.rename(columns=column_mapping)
    return renamed_dataset
```

```python
missing_migrants_df = rename_columns(missing_migrants_df, column_mapping)
missing_migrants_df_2 = rename_columns(missing_migrants_df_2, column_mapping)
missing_migrants_df_3 = rename_columns(missing_migrants_df_3, column_mapping)

create_unique_columns_dataframe([missing_migrants_df, missing_migrants_df_2, missing_migrants_df_3])
```

```python
def count_values_across_dataframes(column_name, dataframes):
    unique_values = []
    
    # Collect unique values from all dataframes
    for df in dataframes:
        unique_values.extend(df[column_name].unique())
    
    print("Unique Values:", unique_values)  # Print the unique values
    
    unique_values = list(set(unique_values))  # Deduplicate the values
    unique_values = sorted(unique_values, key=str)  # Sort the values as strings
    
    result_df = pd.DataFrame(index=unique_values)
    
    # Fill in the dataframe with counts of the values in each dataframe
    for idx, df in enumerate(dataframes, start=1):
        value_counts = df[column_name].value_counts()
        result_df[f"Dataset {idx}"] = result_df.index.map(value_counts).fillna(0)
    
    # Convert the columns to integer data type
    result_df = result_df.astype(int)
    
    return result_df
```

## Finding duplicate incidents in 3 datasets

### Reason: 
Duplicates can distort analytical results, so they need to be identified and handled appropriately.

### Operations:
- Identify duplicates based on a subset of columns.
- Drop duplicates to retain unique records.

```python
def split_coordinates_and_convert(df, coordinate_column, lat_column, lon_column, rounded_lat_column, rounded_lon_column):
    df[[lat_column, lon_column]] = df[coordinate_column].str.split(',', expand=True)
    df[lat_column] = pd.to_numeric(df[lat_column], errors='coerce')
    df[lon_column] = pd.to_numeric(df[lon_column], errors='coerce')
    df[rounded_lat_column] = df[lat_column].round(3)
    df[rounded_lon_column] = df[lon_column].round(3)

# Applying the function to each dataset
split_coordinates_and_convert(missing_migrants_df, 'Coordinates', 'Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon')
split_coordinates_and_convert(missing_migrants_df_2, 'Coordinates', 'Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon')
split_coordinates_and_convert(missing_migrants_df_3, 'Coordinates', 'Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon')

# Display the first few rows to verify the correct splitting, conversion, and rounding
missing_migrants_df[['Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon']].head(), missing_migrants_df_2[['Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon']].head(), missing_migrants_df_3[['Lat', 'Lon', 'Rounded_Lat', 'Rounded_Lon']].head()
```

```python
from datetime import datetime

def convert_month_strings_to_numbers(df):
    # Dictionary mapping month representations to numeric values
    month_mapping = {
        'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
        'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
        'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12,
        'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6, 'July': 7,
       'August': 8, 'September': 9, 'October': 10, 'November':11, 'December': 12
    }
    
    # Convert 'Reported Month' column to numeric month representations
    df['Reported Month'] = df['Reported Month'].map(month_mapping)
    
    return df
```
